```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
Names =[]



for i in soup.find_all('div',class_="presidentListing"):

    Names.append(i.get_text().strip("\n").split("(")[0])

Names
```

```python
Term_of_Office = []

for i in soup.find_all('div',class_="presidentListing"):
    Term_of_Office.append(i.get_text().strip('\n').split("\n")[1])
    
Term_of_Office
```

```python
Former_Presidents = pd.DataFrame()
Former_Presidents['Names'] = Names
Former_Presidents['Term_of_Office'] = Term_of_Office

Former_Presidents
```
