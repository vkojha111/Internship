```python
Product_URL = []
start = 0
end = 2
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        Product_URL.append(i.get_attribute("href"))
    next_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_button.click()
    time.sleep(5)
```

```python
len('Product_URL')
```

```python
import pandas as pd
df=pd.DataFrame()
df['Product_URL']=Product_URL
```

```python
Product_URL
```

```python
Brand = []
Product = []
Price = []
Return_Exchange = []
Expected_Delivery = []
Availability = []
```

```python
for url in Product_URL:
    driver.get(url)
    time.sleep(2)
    
    try:
        brand = driver.find_element(By.XPATH,'//span[@class="a-size-base po-break-word"]')
        Brand.append(brand.text)
    except NoSuchElementException:
            Brand.append('-')
            
    try:
        product = driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        Product.append(product.text)
    except NoSuchElementException:
            Product.append('-')
            
    try:
        price = driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        Price.append(price.text)
    except NoSuchElementException:
            Price.append('-')
            
    try:
        return_exchange = driver.find_element(By.XPATH,'//*[@id="RETURNS_POLICY"]/span/div[2]/span')
        Return_Exchange.append(return_exchange.text)
    except NoSuchElementException:
            Return_Exchange.append('-')
            
    try:
        expect_deliv = driver.find_element(By.XPATH,'//span[@acfv="a-text-bold"]')
        Expected_Delivery.append(expect_deliv.text)
    except NoSuchElementException:
            Expected_Delivery.append('-')
            
    try:
        availability = driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        Availability.append(availability.text)
    except NoSuchElementException:
            Availability.append('-')
```

```python
df = pd.DataFrame()
df['Brand'] = Brand
df['Product'] = Product
df['Price'] = Price
df['Return_Exchange'] = Return_Exchange
df['Expected_Delivery'] = Expected_Delivery
df['Availability'] = Availability
```

```python
driver.close()
```

```python
Hostel_Name = []
Distance = []
Ratings = []
Total_Reviews = []
Private_Price = []
Droms_Price = []
Description = []
```

```python
Hostel_name = driver.find_elements(By.XPATH,'//div[@class="property-name"]')
for i in Hostel_name:
    host_name = i.text
    Hostel_Name.append(host_name)
    
rating = driver.find_elements(By.XPATH,'//div[@class="score-row"]')
for i in rating:
    ratings = i.text
    Ratings.append(ratings)
    
distance = driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
for i in distance:
    Dist = i.text
    Distance.append(Dist)
    
Revews = driver.find_elements(By.XPATH,'//span[@class="left-margin"]')
for i in Revews:
    total_reviews = i.text
    Total_Reviews.append(total_reviews)
    
private = driver.find_elements(By.XPATH,'//strong[1][@class="current"]')
for i in private:
    private_price = i.text
    Private_Price.append(private_price)
    
Droms = driver.find_elements(By.XPATH,'//div[2][@class="property-accommodation-price"]')
for i in Droms:
    drom_price = i.text
    Droms_Price.append(drom_price)
    
discription = driver.find_elements(By.XPATH,'//div[@class="tag-text"]')
for i in discription:
    Disc = i.text
    Description.append(Disc)
```
