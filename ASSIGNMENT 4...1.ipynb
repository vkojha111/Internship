```python
import pandas as pd
import re
import selenium
from selenium import webdriver

from bs4 import BeautifulSoup
import requests

import time

from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait

import warnings
warnings.filterwarnings('ignore')
```

```python
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome('Chromedriver.exe')

driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
Rank = []
Name = []
Artist = []
Views = []
Date = []
```

```python
try:
    all_rank = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
    for i in all_rank:
        rank = i.text
        Rank.append(rank)
except NoSuchElementException:
    Rank.append('-')
    
try:
    all_name = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
    for i in all_name:
        name = i.text
        Name.append(name)
except NoSuchElementsException:
    Artist.append('-')
        
try:
    arti = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
    for i in arti:
        artist = i.text
        Artist.append(artist)
except NoSuchElementsException:
    Artist.append('-')
    
try:
    view = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
    for i in view:
        views = i.text
        Views.append(views)
except NoSuchElementException:
    Views.append('-')
    

try:
    dat = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
    for i in dat:
        date = i.text
        Date.append(date)
except NoSuchElementException:
    Date.append('-')
```

```python
df = pd.DataFrame()
df['Rank'] = Rank
df['Name'] = Name
df['Artist'] = Artist
df['Views(In Billions)'] = Views
df['Date'] = Date
```

```python
df.head()
```

```python
driver.close()
```

# Soluteion No:8

```python
driver = webdriver.Chrome('Chromedriver.exe')
```

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
view_dataset = driver.find_element(By.XPATH,'//div[@class="flex flex-wrap justify-center gap-5"][1]/a')
view_dataset.click()
```

```python
URL=[]
```

```python
url = driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for i in url:
    page_url = i.get_attribute("href")
    URL.append(page_url)
```

```python
URL
```

```python
Dataset_name = []
Data_type = []
Task = []
Attribute_type = []
No_of_instances = []
No_of_attributes = []
Year = []
```

```python
for urls in URL:
    driver.get(urls)
    
    for i in driver.find_elements(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]'):
        Dataset_name.append(i.text)
        
    for i in driver.find_elements(By.XPATH,'//div[@class="relative flex flex-col gap-4 bg-base-100 p-4 shadow"][1]/div'):
        Data_type.append(i.text)
        
    for i in driver.find_elements(By.XPATH,'//div[@class="col-span-4"][3]/p[1]'):
        Task.append(i.text)
        
    for i in driver.find_elements(By.XPATH,'//div[@class="col-span-4"][4]/p[1]'):
        Attribute_type.append(i.text)
        
    for i in driver.find_elements(By.XPATH,'//div[@class="col-span-4"][5]/p[1]'):
        No_of_instances.append(i.text)
        
    for i in driver.find_elements(By.XPATH,'//div[@class="col-span-4"][6]/p[1]'):
        No_of_attributes.append(i.text)
        
    for i in driver.find_elements(By.XPATH,'//h2[@class="text-primary-content"]'):
        Year.append(i.text)
```

```python
df = pd.DataFrame()
df['Dataset_name'] = Dataset_name
df['Data_type'] = Data_type
df['Task'] = Task
df['Attribute_type'] = Attribute_type
df['No_of_instances'] = No_of_instances
df['No_of_attributes'] = No_of_attributes
df['Year'] = Year
```

```python
driver.close()
```

# Solution No: 9

```python
driver = webdriver.Chrome('Chromedriver.exe')
```

```python
driver.get('https://www.naukri.com/hr-recruiters-consultants')
```

```python
Name = []
Designation = []
Company = []
Skills_they_hire_for = []
Location = []
```

```python
name = driver.find_elements(By.XPATH,'//span[@class="fl ellipsis"]')
for i in name:
    names = i.text
    Name.append(names)
    
designation = driver.find_elements(By.XPATH,'//span[@class="ellipsis clr"]')
for i in designation:
    desig = i.text
    Designation.append(desig)
    
company = driver.find_elements(By.XPATH,'//small[@class="ellipsis"]')
for i in company:
    com = i.text
    Company.append(com)
    
skill = driver.find_elements(By.XPATH,'//div[@class="hireSec"]')
for i in skill:
    skills = i.text
    Skills_they_hire_for.append(skills)
    
location = driver.find_elements(By.XPATH,'//div[@class="vcard"]//p[1]/span/small')
for i in location:
    loc = i.text
    Location.append(loc)
```

```python
print(len('Name'),len('Designation'),len('Company'),len('Skills_they_hire_for'),len('Location'))
```

```python
df = pd.DataFrame()
df['Name'] = Name[0:5]
df['Designation'] = Designation[0:5]
df['Company'] = Company[0:5]
df['Skills_they_hire_for'] = Skills_they_hire_for[0:5]
df['Location'] = Location[0:5]
```
