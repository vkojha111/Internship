```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.imdb.com/list/ls058711268/')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
Name = []

for i in soup.find_all('h3',class_="lister-item-header"):
    Name.append(i.get_text().strip('\n').split('\n')[1])
    
Name
```

```python
Year = []

for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    Year.append(i.text)
    


Year
```

```python
Ratings = []
for i in soup.find_all('div',class_="ipl-rating-star small"):
    Ratings.append(i.get_text().strip("\n").split("\n")[0])
    
Ratings
```

```python
import pandas as pd
```

```python
IMDB_Rating = pd.DataFrame()
IMDB_Rating['Name'] = Name
IMDB_Rating['Year'] = Year
IMDB_Rating['Ratings'] = Ratings
IMDB_Rating
```
